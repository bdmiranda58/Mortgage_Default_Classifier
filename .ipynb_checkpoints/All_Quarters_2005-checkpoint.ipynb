# Fannie Mae Loan Default Classifier

```python
import pandas as pd
```

```python
acq_fp = './data/acquisition_data_dict_summary.csv'
list_acq_col_names = pd.read_csv(acq_fp)['Field Name'].tolist()
```

```python
perf_fp = './data/performance_data_dict_summary.csv'
list_perf_col_names = pd.read_csv(perf_fp)['Field Name'].tolist()
```

```python
df_acq = pd.DataFrame()

for i in range(1,5):
    file = f'./data/2005Q{i}/Acquisition_2005Q{i}.txt'
    df_temp = pd.read_csv(file, sep="|", header=None, names = list_acq_col_names)
    df_acq = pd.concat([df_acq, df_temp])

df_acq.head()
```

```python
df_perf = pd.DataFrame()

for i in range(1,5):
    file = f'./data/2005Q{i}/Performance_2005Q{i}.txt'
    df_temp = pd.read_csv(file, sep="|", header=None, names = list_perf_col_names)
    df_perf = pd.concat([df_perf, df_temp])

df_perf.head()
```
